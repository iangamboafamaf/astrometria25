```python
import mi_funcion as mf
import numpy as np
import matplotlib.pyplot as plt
```

```python
numeros = mf.congruencial_lineal(1000,142)
```

```python
x = numeros[1:]
y = numeros[:-1]
plt.plot(x,y,'o')
plt.xlabel('$x_i$')
plt.ylabel('$x_{i-1}$')
plt.title('Generador Congruencial Lineal')
plt.axis('on')
plt.grid()
plt.show()
```

```python
N_caminatas = 10
N_pasos = 1000
```

```python
x = np.zeros((N_caminatas,N_pasos))
y = np.zeros((N_caminatas,N_pasos))
# N caminatas filas y N pasos columnas
for i in range(N_caminatas):
    for j in range(1, N_pasos):
        salto = numeros[(i*N_pasos + j) % len(numeros)] * 2 * np.sqrt(2) - np.sqrt(2)
        x[i, j] = x[i, j-1] + salto
        salto = numeros[(i*N_pasos + j + 1) % len(numeros)] * 2 * np.sqrt(2) - np.sqrt(2)
        y[i, j] = y[i, j-1] + salto
        
    
```

```python
for i in range(N_caminatas):
    plt.plot(x[i,:],y[i,:])
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'{N_caminatas} Caminatas Aleatorias de {N_pasos} pasos')
```

```python
distancias = np.sqrt(x**2 + y**2)
```

```python
for i in range(N_caminatas):
    plt.plot(distancias[i])
plt.plot(np.mean(distancias,axis=0), 'k', lw=2)
```

```python
distancias = np.sqrt(x[:,-1]**2 + y[:,-1]**2)
print(f'Distancias promedio: {np.mean(distancias):7.2f} +/- {np.std(distancias):7.2f}')
```
